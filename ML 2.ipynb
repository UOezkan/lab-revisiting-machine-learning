```python
num = data.select_dtypes(np.number)
num
```

```python
df = pd.DataFrame(num.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
```

```python
#replace all the none values with the mean 

def replace_null_with_mean(df):
    for column in df.columns:
        mean_value = df[column].mean()
        df[column].fillna(mean_value, inplace=True)
    return df

# Example usage:
cleaned_num = replace_null_with_mean(num)
cleaned_num
```
